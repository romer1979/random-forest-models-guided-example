```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
y2015 = pd.read_csv(
    'https://www.dropbox.com/s/0so14yudedjmm5m/LoanStats3d.csv?dl=1',
    skipinitialspace=True,
    header=1
)
```

```python
y2015.describe()
```

This is a big data frame with 85 columns and over 400,000 rows

```python
y2015.tail()
```

Last 2 rows need to be removed from the data frame

```python
y2015 = y2015[:-2]
```

Checking the distinct counts of the categorical variables

```python
categorical = y2015.select_dtypes(include=['object'])
for i in categorical:
    column = categorical[i]
    print(i)
    print(column.nunique())
```

Dropping all the columns with more than 10 distinct counts. plus changing "int_rate" feature to numeric since it can not be categorical

```python
y2015['int_rate'] = pd.to_numeric(y2015['int_rate'].str.strip('%'), errors='coerce')

# Drop other columns with many unique variables
y2015.drop(['url', 'emp_title', 'zip_code', 'earliest_cr_line', 'revol_util', 'sub_grade', 'addr_state', 'desc', 
            'emp_length', 'issue_d', 'purpose', 'title', 'last_pymnt_d', 'next_pymnt_d', 'last_credit_pull_d'], 
           1, inplace=True)
```

Dropping all non important features and related ones

```python
y2015.drop(['id', 'member_id', 'funded_amnt_inv',
       'mths_since_last_delinq',
       'mths_since_last_record', 
       'out_prncp_inv',
       'total_pymnt_inv', 'total_rec_prncp', 'total_rec_int',
       'total_rec_late_fee',
       'dti_joint', 'verification_status_joint',
       'open_acc_6m',
       'open_il_6m', 'open_il_12m', 'open_il_24m',
       'il_util', 'open_rv_12m', 'open_rv_24m',
       'all_util', 'total_rev_hi_lim', 'inq_fi', 'inq_last_12m',
       'acc_open_past_24mths', 'bc_open_to_buy', 'bc_util',
       'mo_sin_old_il_acct',
       'mo_sin_old_rev_tl_op', 'mo_sin_rcnt_rev_tl_op', 'mo_sin_rcnt_tl',
       'mths_since_recent_bc_dlq',
       'mths_since_recent_inq', 'mths_since_recent_revol_delinq',
       'num_accts_ever_120_pd', 'num_actv_bc_tl', 'num_actv_rev_tl',
       'num_bc_sats', 'num_bc_tl', 'num_il_tl', 'num_op_rev_tl',
       'num_tl_120dpd_2m',
       'num_tl_30dpd', 'num_tl_90g_dpd_24m', 'num_tl_op_past_12m',
       'pct_tl_nvr_dlq', 'percent_bc_gt_75'], 
           1, inplace=True)
```

get dummy columns for all remaining categorical features

```python
pd.get_dummies(y2015)
```

```python
from sklearn import ensemble
from sklearn.model_selection import cross_val_score

rfc = ensemble.RandomForestClassifier()
X = y2015.drop('loan_status', 1)
Y = y2015['loan_status']
X = pd.get_dummies(X)
X = X.dropna(axis=1)

print(cross_val_score(rfc, X, Y, cv=10))
print(cross_val_score(rfc, X, Y, cv=10).mean())
```

The model performed well after dropping almost 50% of the features included in the original model. average R-Square for this model is more than 0.9 and very close to the original model average R-Square of 0.97

Now lets drop the features related to payment amount and outstanding principal and check the accuracy of the model

```python
y2015.drop(['out_prncp', 'total_pymnt','last_pymnt_amnt'], 1, inplace=True)
```

```python
from sklearn import ensemble
from sklearn.model_selection import cross_val_score

rfc = ensemble.RandomForestClassifier()
X = y2015.drop('loan_status', 1)
Y = y2015['loan_status']
X = pd.get_dummies(X)
X = X.dropna(axis=1)

print(cross_val_score(rfc, X, Y, cv=10))
print(cross_val_score(rfc, X, Y, cv=10).mean())
```
